```python
# Missing Values Analysis

# Count missing values and percentage per column
missing_count = df.isnull().sum()
missing_pct   = (missing_count / len(df)) * 100

# Build summary table (only columns that actually have missing values)
missing_df = pd.DataFrame({
    'Missing Values': missing_count,
    'Percentage (%)' : missing_pct
})
missing_df = missing_df[missing_df['Missing Values'] > 0].sort_values('Missing Values', ascending=False)

print("Columns with Missing Values:")
print(missing_df.to_string())
print()
print(f"Column with the most missing data: '{missing_df.index[0]}'  "
      f"({missing_df['Missing Values'].iloc[0]:,} missing  |  "
      f"{missing_df['Percentage (%)'].iloc[0]:.2f}%)")

#Visualization ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left: Missing value counts
axes[0].bar(missing_df.index, missing_df['Missing Values'], color='steelblue', edgecolor='black')
axes[0].set_title('Missing Values Count', fontsize=13)
axes[0].set_xlabel('Column')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# Right: Missing value percentages
axes[1].bar(missing_df.index, missing_df['Percentage (%)'], color='salmon', edgecolor='black')
axes[1].set_title('Missing Values Percentage', fontsize=13)
axes[1].set_xlabel('Column')
axes[1].set_ylabel('Percentage (%)')
axes[1].tick_params(axis='x', rotation=45)

plt.suptitle('Missing Data Overview', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Drop Irrelevant Columns 

# These columns don't add value to analysis:
# page_url   : just a web link
# location_id: numeric ID, we already have location name

cols_to_drop = ['page_url', 'location_id']

df = df.drop(columns=cols_to_drop)

# Confirm result
print(f"Dropped columns : {cols_to_drop}")
print(f"Remaining columns ({df.shape[1]}): {df.columns.tolist()}")
print(f"New shape: {df.shape}")
```

```python
df.info()
```

```python
# Imputation Strategy 

# Separate columns by type
categorical_cols = df.select_dtypes(include='object').columns.tolist()
numerical_cols   = df.select_dtypes(include='number').columns.tolist()

# Fill categorical nulls with "Unknown"
df[categorical_cols] = df[categorical_cols].fillna('Unknown')

# Fill numerical nulls with column mean
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].mean())

# Confirm no missing values remain
print("Missing values after imputation:")
print(df.isnull().sum())
print()
print(f"Total missing values: {df.isnull().sum().sum()}")
```

```python
# Verification: Check for Remaining Missing Values

total_missing = df.isnull().sum().sum()

if total_missing == 0:
    print("✅ Imputation successful! No missing values remain in the dataset.")
else:
    print(f"⚠️ {total_missing} missing values still remain:")
    print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# --- Safe conversion to Marla ---

def convert_to_marla(area_str):
    try:
        parts = str(area_str).strip().split()
        value = float(parts[0])
        unit  = parts[1].lower()

        if unit == 'kanal':
            return value * 20
        elif unit == 'marla':
            return value
        else:
            return None   # unknown unit → flag it for review

    except:
        return None       # catches any malformed values

df['area_in_marla'] = df['area'].apply(convert_to_marla)

# Check how many couldn't be converted
failed = df['area_in_marla'].isnull().sum()
print(f"Successfully converted: {len(df) - failed}")
print(f"Failed to convert    : {failed}")

if failed > 0:
    print("\nUnconverted area values:")
    print(df[df['area_in_marla'].isnull()]['area'].value_counts())
```

```python
# conversion to Marla 

def convert_to_marla(area_str):
    try:
        parts = str(area_str).strip().split()
        value = float(parts[0].replace(',', ''))  # remove commas e.g. "1,000" → "1000"
        unit  = parts[1].lower()

        if unit == 'kanal':
            return value * 20
        elif unit == 'marla':
            return value
        else:
            return None

    except:
        return None

df['area_in_marla'] = df['area'].apply(convert_to_marla)

# Verify
print("Sample conversion check:")
print(df[['area', 'area_in_marla']].head(10))
print()
print(f"Failed to convert : {df['area_in_marla'].isnull().sum()}")
```

## Phase:3 Outlier Detection

```python
#Outlier Detection: price & area_in_marla ---

# Box plots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

df.boxplot(column='price',         ax=axes[0])
df.boxplot(column='area_in_marla', ax=axes[1])

axes[0].set_title('Price')
axes[1].set_title('Area in Marla')

plt.suptitle('Outlier Visualization', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# --- IQR Outlier Count ---

for col in ['price', 'area_in_marla']:
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower    = Q1 - 1.5 * IQR
    upper    = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col} → Outliers: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)")
```

```python
#Visual Check: Spot Unrealistic Values ---

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Boxplot — Price
axes[0].boxplot(df['price'], patch_artist=True,
                boxprops=dict(facecolor='steelblue', color='black'))
axes[0].set_title('Price Distribution')
axes[0].set_ylabel('Price (PKR)')
axes[0].set_xticks([])

# 2. Boxplot — Area in Marla
axes[1].boxplot(df['area_in_marla'], patch_artist=True,
                boxprops=dict(facecolor='salmon', color='black'))
axes[1].set_title('Area Distribution')
axes[1].set_ylabel('Area (Marla)')
axes[1].set_xticks([])

# 3. Scatter plot — Price vs Area
axes[2].scatter(df['area_in_marla'], df['price'], alpha=0.3, s=5, color='steelblue')
axes[2].set_title('Price vs Area')
axes[2].set_xlabel('Area (Marla)')
axes[2].set_ylabel('Price (PKR)')

plt.suptitle('Visual Check for Unrealistic Values', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# --- Print unrealistic value counts ---
zero_price = (df['price'] == 0).sum()
zero_area  = (df['area_in_marla'] == 0).sum()

print(f"Properties with price = 0  : {zero_price}")
print(f"Properties with area  = 0  : {zero_area}")
```

```python
# 1.Price Distribution: Histogram AND KDE 

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# 1. Original price
sns.histplot(df['price'], kde=True, ax=axes[0], color='steelblue')
axes[0].set_title('Price Distribution (Original)')
axes[0].set_xlabel('Price (PKR)')
axes[0].set_ylabel('Count')

# 2. Log-transformed price (filter out 0s to avoid log(0) error)
log_price = df[df['price'] > 0]['price'].apply(np.log)
sns.histplot(log_price, kde=True, ax=axes[1], color='salmon')
axes[1].set_title('Price Distribution (Log Scale)')
axes[1].set_xlabel('log(Price)')
axes[1].set_ylabel('Count')

plt.suptitle('Price Distribution Analysis', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# --- Skewness check ---
skewness = df['price'].skew()
print(f"Skewness of price: {skewness:.2f}")

if skewness > 1:
    print("→ Highly right-skewed. Log transformation recommended.")
elif skewness > 0:
    print("→ Slightly right-skewed.")
else:
    print("→ Approximately normal or left-skewed.")
```

## B. Bivariate Analysis (Two Variables)

```python
#5 Price vs Top 3 Cities (Boxplot) ---

# Get top 3 cities
top3_cities = df['city'].value_counts().head(3).index.tolist()

# Filter dataset to top 3 cities only
df_top3 = df[df['city'].isin(top3_cities)]

# Prepare data for each city separately
city_prices = [df_top3[df_top3['city'] == city]['price'] for city in top3_cities]

# Boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(city_prices, labels=top3_cities, patch_artist=True)

plt.title('Price Distribution — Top 3 Cities', fontsize=14, fontweight='bold')
plt.xlabel('City')
plt.ylabel('Price (PKR)')
plt.tight_layout()
plt.show()

# Print median price per city
print("Median Price per City:")
for city in top3_cities:
    median = df_top3[df_top3['city'] == city]['price'].median()
    print(f"  {city:<15} PKR {median:>15,.0f}")
```

```python
#6 Price vs Area (Scatter Plot) ---

# Filter out extreme values for a clearer view
df_filtered = df[(df['area_in_marla'] <= 500) & (df['price'] <= 200_000_000)]

plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['area_in_marla'], df_filtered['price'],
            alpha=0.3, s=5, color='steelblue')

plt.title('Price vs Area in Marla', fontsize=14, fontweight='bold')
plt.xlabel('Area (Marla)')
plt.ylabel('Price (PKR)')
plt.tight_layout()
plt.show()

# --- Spot suspicious properties: small area but very high price ---
suspicious = df[(df['area_in_marla'] <= 5) & (df['price'] >= 50_000_000)]

print(f"Small area (≤5 Marla) but high price (≥50M PKR): {len(suspicious)} properties")
print()
print(suspicious[['property_type', 'area_in_marla', 'price', 'city', 'location']].head(10))
```

```python
# Bathrooms vs Price

# Filter to realistic bath counts only (remove outliers like 403 baths)
df_baths = df[df['baths'] <= 10]

# Calculate median price per bathroom count
median_price_by_baths = df_baths.groupby('baths')['price'].median()

# Bar chart — median price per bathroom count
plt.figure(figsize=(10, 5))
plt.bar(median_price_by_baths.index, median_price_by_baths.values,
        color='steelblue', edgecolor='black')

plt.title('Median Price by Number of Bathrooms', fontsize=14, fontweight='bold')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Median Price (PKR)')
plt.xticks(median_price_by_baths.index)
plt.tight_layout()
plt.show()

# Print the values
print("Median Price by Bathrooms:")
for baths, price in median_price_by_baths.items():
    print(f"  {baths} bath(s)  →  PKR {price:>15,.0f}")
```

```python
# Date vs Count: Listings per Year

# Convert date_added to datetime and extract year
df['date_added'] = pd.to_datetime(df['date_added'], format='%m-%d-%Y', errors='coerce')
df['year'] = df['date_added'].dt.year

# Count listings per year
listings_per_year = df['year'].value_counts().sort_index()

# Line chart
plt.figure(figsize=(10, 5))
plt.plot(listings_per_year.index, listings_per_year.values,
         marker='o', color='steelblue', linewidth=2)

plt.title('Number of Listings per Year', fontsize=14, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Number of Listings')
plt.xticks(listings_per_year.index)
plt.show()

# Print result
print("Listings per Year:")
for year, count in listings_per_year.items():
    print(f"  {int(year)}  →  {count:,} listings")
```

# 🏢 Real Estate Market Findings & Business Answers

---

## 1. Which city shows the highest growth in high-value properties over the years?

Karachi shows the highest volume of growth in high-value properties defined as >50 Million PKR.

* High value listings in Karachi increased from 300 in 2018 to over 5,300 in 2019.
* Lahore also showed massive growth, increasing from 150 to over 4,200 listings.

This indicates aggressive premium real estate development or a shift in high end listings to digital platforms during that period.

---

## 2. What are the most popular property types in each city, and how has this preference changed?

Property preferences are highly regional:

* **Karachi:** Flats are the #1 most listed property, slightly ahead of houses. This reflects high population density and vertical expansion.
* **Lahore, Islamabad & Rawalpindi:** Houses dominate the market.
* **Faisalabad & Lahore:** Upper portion rentals are also popular, showing a strong culture of renting portions of houses instead of purpose-built flats.

---

## 3. How do property prices vary with the number of bedrooms?

Prices increase up to a point and then plateau:

* Prices rise sharply from 1 to 7 bedrooms.
* Example: 2-bed median =3M PKR, 7-bed median = 29M PKR.
* Beyond 7 bedrooms, prices plateau or slightly decline (= 23M–26M PKR).

This shows that very large properties are often older or used for commercial/multi-family purposes instead of luxury homes.

---

Top active agencies in Lahore, Karachi, and Islamabad include:

* Mash Allah Estate & Builders

## 5. Is there a seasonal trend in property listings?

Yes, a strong seasonal pattern exists:

* Winter/Spring: 2,000 – 5,000 listings/month
* June: 50,000 listings
* July: 88,000 listings (peak)

This suggests high market activity in mid-year, likely due to fiscal cycles, school holidays, and new project launches.

## 6. How do property sizes correlate with prices across cities?

Correlation between area_in_marla and price is very weak:

* Karachi: 0.10
* Lahore: 0.03
* Islamabad: 0.02

**Key Insight:** Location matters far more than size. A small property in DHA or Gulberg can cost significantly more than a large property in rural areas.

---

## 7. Which property types are prevalent in certain provinces and impact prices?

* **Sindh:** Dominated by flats → lower median entry prices, high density living.
* **Punjab:** Dominated by houses and upper portions → higher baseline property prices.
* **Islamabad Capital:** Balanced mix of houses and flats for both permanent and transient residents.

---

## 8. How does the number of bathrooms influence property prices?

Bathroom trends show a similar pattern to bedrooms:

* Prices increase from 1 bath (=1.6M PKR) to 8 baths (=35M PKR).
* At 9+ baths, median price drops to 25M PKR.

Extremely high bathroom counts usually indicate commercial or multi-unit properties rather than luxury homes.

Market distribution varies by city:
